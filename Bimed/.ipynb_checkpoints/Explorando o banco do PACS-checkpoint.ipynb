```python
import psycopg2
import pandas as pd
conn = psycopg2.connect("host=pacs.lais.huol.ufrn.br dbname=pacsdb user=jailton password=l41sjailton")
cur = conn.cursor()
```

```python
meses = range(1,13)
anos = range(2013,2017)
modalidades = ( "RTSTRUCT", "US", "CR", "MR", "CT",
                "XA", "SC", "PR", "OT", "PDF", "PT",
                "XC", "SR", "KO", "ECG", "DX")

estudo = {}
```

```python
for ano in anos:
    if ano not in estudo:
        estudo[ano] = {}
        dados_ano = estudo[ano]

    for mes in meses:
        quantidades_mes = []
        for modalidade in modalidades:
            sql = """
            SELECT COUNT(*) from series
            WHERE extract(year from created_time) = %d AND extract(month from created_time) = %d
            AND modality LIKE '%s'
            """ % (ano, mes, "%"+modalidade+"%")
            cur.execute(sql)
            quantidades_mes.append(cur.fetchall()[0][0])
        dados_ano[mes] = pd.Series(quantidades_mes, index=modalidades)
```
