```python
%matplotlib inline
import psycopg2
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
meses = ('Janeiro', 'Fevereiro', u'Mar√ßo', 'Abril', 
         'Maio', 'Junho', 'Julho', 'Agosto',
         'Setembro', 'Outubro', 'Novembro', 'Dezembro')
meses_min = ('Jan', 'Fev', u'Mar', 'Abr', 
         'Mai', 'Jun', 'Jul', 'Ago',
         'Set', 'Out', 'Nov', 'Dez')

ano2013 = pd.read_csv("ano2013.csv", index_col="Modalidade")
ano2014 = pd.read_csv("ano2014.csv", index_col="Modalidade")
ano2015 = pd.read_csv("ano2015.csv", index_col="Modalidade")
ano2016 = pd.read_csv("ano2016.csv", index_col="Modalidade")
```

<h1>Numeros espec√≠ficos</h1>

```python
outros = ('RTSTRUCT','US','XA','SC','PR','OT','PDF','PT','XC','SR','KO','ECG','DX')
```

<h2>2016</h2>

```python
outros_dados2016 = []
for i in range(len(meses)):
    dado_acumulado = 0
    for exame in outros:
        dado_acumulado+= ano2016.loc[exame][i]
    outros_dados2016.append(dado_acumulado)
outros_dados2016

# Create the general blog and the "subplots" i.e. the bars
f, ax1 = plt.subplots(1, figsize=(8,3))

# Set the bar width
bar_width = 0.4

# positions of the left bar-boundaries
bar_l = [i+1 for i in range(len(list(ano2016)[:-1]))] 

# positions of the x-axis ticks (center of the bars as bar labels)
tick_pos = [i+(bar_width/2) for i in bar_l] 

# Create a bar plot, in position bar_1
ax1.bar(bar_l, 
        # using the pre_score data
        outros_dados2016, 
        # set the width
        width=bar_width,
        # with the label pre score
        label='Outros', 
        # with alpha 0.5
        alpha=0.75, 
        # with color
        color='#6BB9F0')

# Create a bar plot, in position bar_1
ax1.bar(bar_l, 
        # using the mid_score data
        ano2016.loc['CR'][:-1], 
        # set the width
        width=bar_width,
        # with pre_score on the bottom
        bottom=outros_dados2016, 
        # with the label mid score
        label='Radiografia', 
        # with alpha 0.5
        alpha=0.75, 
        # with color
        color='#F1911E')

# Create a bar plot, in position bar_1
ax1.bar(bar_l, 
        # using the post_score data
        ano2016.loc['MR'][:-1], 
        # set the width
        width=bar_width,
        # with pre_score and mid_score on the bottom
        bottom=[i+j for i,j in zip(ano2016.loc['CR'][:-1],outros_dados2016)], 
        # with the label post score
        label=u'Resson√¢ncia magn√©tica', 
        # with alpha 0.5
        alpha=0.75, 
        # with color
        color='#F1BD1A')

ax1.bar(bar_l, 
        # using the post_score data
        ano2016.loc['CT'][:-1], 
        # set the width
        width=bar_width,
        # with pre_score and mid_score on the bottom
        bottom=[i+j+k for i,j,k in zip(ano2016.loc['CR'][:-1],ano2016.loc['MR'][:-1],outros_dados2016)], 
        # with the label post score
        label='Tomografia computadorizada', 
        # with alpha 0.5
        alpha=0.75, 
        # with color
        color='#F4561D')

# set the x ticks with names
plt.xticks(tick_pos, meses_min)

# Set the label and legends
ax1.set_ylabel("Quantidade de exames")
ax1.set_xlabel("Meses do ano de 2016")
plt.legend(loc='upper left')

box = ax1.get_position()
ax1.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Set a buffer around the edge
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
```
